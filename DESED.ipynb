```python
# === Step 0: Install system dependencies and packages ===
!apt-get update
!apt-get install -y sox libsox-dev libsox-fmt-all ffmpeg

!pip install torch torchaudio librosa matplotlib numpy tqdm pyyaml sed_eval psds_eval desed scaper

# === Step 1: Upload audio ===
from google.colab import files
uploaded = files.upload()

# Pick the first uploaded file
audio_file = list(uploaded.keys())[0]
print("Uploaded file:", audio_file)

# === Step 2: Convert to .wav using ffmpeg ===
import os
base_name = "test_audio.wav"
os.makedirs("/content/dcase20_task4/dataset/audio/validation", exist_ok=True)

!ffmpeg -y -i "{audio_file}" "/content/dcase20_task4/dataset/audio/validation/{base_name}"

# === Step 3: Create dummy metadata TSV ===
os.makedirs("/content/dcase20_task4/dataset/metadata/validation", exist_ok=True)
metadata_path = "/content/dcase20_task4/dataset/metadata/validation/validation.tsv"
with open(metadata_path, "w") as f:
    f.write("filename\tonset\toffset\tevent_label\n")
    f.write(f"{base_name}\t0\t1\tdummy\n")

# === Step 4: Download checkpoint if missing ===
checkpoint_path = "/content/dcase20_task4/crnn.pth"
if not os.path.exists(checkpoint_path):
    !wget https://zenodo.org/record/3826813/files/crnn.pth?download=1 -O "{checkpoint_path}"

# === Step 5: Run inference ===
%cd /content/dcase20_task4/baseline
!mkdir -p stored_data/baseline

!python TestModel.py \
  -m "../crnn.pth" \
  -g ../dataset/metadata/validation/validation.tsv \
  -ga ../dataset/audio/validation \
  -s stored_data/baseline/validation_predictions.tsv

# === Step 6: Read predictions and print events ===
import pandas as pd

preds = pd.read_csv('stored_data/baseline/validation_predictions.tsv', sep='\t')
print("\nDetected events:")
for idx, row in preds.iterrows():
    print(f"{row['filename']} | {row['onset_time']}s → {row['offset_time']}s | {row['event_label']} | confidence {row['confidence']:.2f}")
```
